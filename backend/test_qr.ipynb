## 1. This is the unique token we will eventually save in the MySQL database

```python
import qrcode
user_token = "USR-RAJTILAK-998877"
```

```python
# This will hdecide the look and size of the QR code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
```

```python
# Add the token data in the QR code
qr.add_data(user_token)
qr.make(fit=True)
```

```python
# Generate the image and save it to folder
img = qr.make_image(fill_color="black",back_color="white")
img.save("rajtilak_id.png")

print(f"Success! QR Code generated for token: {user_token}")
```
